```python
import json
with open('/content/DataEngineeringQ2.json', 'r') as file:
    data = json.load(file)
# data
def transform_gen(gender):
    if gender == 'M':
        return 'male'
    elif gender == 'F':
        return 'female'
    else:
        return 'null'
data1 = []
for item in data:
    data1.append({
        'appointmentId': item.get('appointmentId'),
        'phoneNumber': item.get('phoneNumber'),
        'firstName': item['patientDetails'].get('firstName'),
        'lastName': item['patientDetails'].get('lastName'),
        'gender': transform_gen(item['patientDetails'].get('gender')),
        'DOB': item['patientDetails'].get('birthDate'),
        'medicines': item['consultationData'].get('medicines', [])})

for entry in data1:
    print(entry)
```

```python
def is_valid_mobile(phone_number):
    # Remove any non-digit characters
    cleaned_number = ''.join(filter(str.isdigit, phone_number))

    # Check if the cleaned number is valid
    if cleaned_number.startswith('91') or cleaned_number.startswith('+91'):
        return 6000000000 <= int(cleaned_number[3:]) <= 9999999999
    return False

for entry in data1:
    entry['isValidMobile'] = is_valid_mobile(entry['phoneNumber'])

# Print the updated data
for entry in data1:
    print(entry)
```

```python
import hashlib

def hash_phone_number(phone_number):
    if is_valid_mobile(phone_number):
        return hashlib.sha256(phone_number.encode()).hexdigest()
    return None

for entry in data1:
    entry['phoneNumberHash'] = hash_phone_number(entry['phoneNumber'])

# Print the updated data
for entry in data1:
    print(entry)
```

```python
import json
import hashlib
from datetime import datetime

# Load JSON data
with open('/content/DataEngineeringQ2.json', 'r') as file:
    data = json.load(file)

# Define a function to validate and hash phone numbers
def validate_and_hash_phone(number):
    valid_prefixes = ('+91', '91')
    if number.startswith(valid_prefixes) and len(number) in (12, 13) and number[-10:].isdigit() and 6000000000 <= int(number[-10:]) <= 9999999999:
        return True, hashlib.sha256(number[-10:].encode()).hexdigest()
    else:
        return False, None

# Define a function to calculate age from DOB
def calculate_age(dob):
    if dob:
        # Adjust the format to match the datetime in your JSON data
        birth_date = datetime.strptime(dob, '%Y-%m-%dT%H:%M:%S.%fZ')
        today = datetime.today()
        return today.year - birth_date.year - ((today.month, today.day) < (birth_date.month, birth_date.day))
    else:
        return None


# Process data
for record in data:
    patient = record['patientDetails']
    consultation = record['consultationData']

    # Transform gender
    gender = patient.get('gender', 'others').upper()
    gender = 'male' if gender == 'M' else 'female' if gender == 'F' else 'others'

    # Rename birthDate to DOB and calculate age
    dob = patient.get('birthDate')
    record['DOB'] = dob
    record['Age'] = calculate_age(dob)

    # Create fullName
    record['fullName'] = f"{patient.get('firstName', '')} {patient.get('lastName', '')}".strip()

    # Validate and hash phone number
    is_valid, phone_hash = validate_and_hash_phone(record['phoneNumber'])
    record['isValidMobile'] = is_valid
    record['phoneNumberHash'] = phone_hash if is_valid else None

    # Aggregate medicine data
    record['noOfMedicines'] = len(consultation['medicines'])
    active_medicines = [med for med in consultation['medicines'] if med.get('IsActive', False)]
    record['noOfActiveMedicines'] = len(active_medicines)
    record['noOfInActiveMedicines'] = record['noOfMedicines'] - record['noOfActiveMedicines']


# Save or process the transformed data as needed
```
